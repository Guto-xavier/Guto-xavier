```python
import requests as r
url = 'https://api.covid19api.com/dayone/country/brazil'
resp = r.get(url)
```

```python
resp.status_code
```

```python
raw_data = resp.json()
raw_data[0]
```

```python
final_data = []
for obs in raw_data:
  final_data.append([obs['Confirmed'] , obs['Deaths'] , obs['Recovered'] , obs['Active'] , obs['Date'] ])
```

```python
final_data
```

```python
final_data.insert(0, ['confirmados' , 'obitos' , 'recuperados' , 'ativos' , 'data'])
final_data
```

```python
confirmados = 0
obitos = 1
recuperados = 2
ativos = 3
data = 4
```

```python
for i in range(1, len(final_data)):
  final_data[i][data] = final_data[i][data][:10]
```

```python
#da forma como o curso Let's Code ensina

def get_datasets(y, labels):
  if type(y[0]) == list:
    datasets =[]
    for i in range(len(y)):
      datasets.append({
        'label' : labels[i] ,
        'data' : y[i]
      })
    return datasets
  else:
    return [
            {
                'label' : labels[0],
                'data' : y
            }
    ]
```

```python
#da forma como o curso Let's Code ensina
def set_title(title= ''):
  if title != '':
    display = 'true'
  else:
    display = 'false'
  return {
      'title' : title ,
      'display' : display

  }
```

```python
#da forma como o curso Let's Code ensina
def create_chart(x, y, labels, kind='bar', title=''):

  datasets = get_datasets(y, labels)
  options = set_title(title)

  chart={
      'type' : kind,
      'data' : {
          'labels' : x,
          'datasts' : datasets
      },
      'options' : options
  }

  return chart
```

```python
def get_api_chart(chart):
  url_base = 'https://quickchart.io/chart'
  resp = r.get(f'{url_base}?c={str(chart)}')
  return resp.content
```

```python
def save_image(path, content):
  with open(path, 'wb') as image:
    image.write(content)
    
```

```python
from PIL import Image
from IPython.display import display
```

```python
def display_image(path):
  img_pill = Image.open(path)
  display(img_pill)
```

```python
y_data_1 = []
for obs in final_data[1::10]:
  y_data_1.append(obs[confirmados])

y_data_2 = []
for obs in final_data[1::10]:
  y_data_2.append(obs[confirmados])

labels =['confirmados' , 'recuperados']

x = []
for obs in final_data[1::10]:
  x.append(obs[data].strftime('%d/%m/%y'))

chart = create_chart(x, [y_data_1 , y_data_2], labels, kind='bar', title='Gráfico confirmados X recuperados')
chart_content = get_api_chart(chart)
save_image('grafico.png' , chart_content)
display_image('grafico.png')
```
