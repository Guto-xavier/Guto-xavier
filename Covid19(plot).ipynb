#2 - tentaiva de definir funes para plot, pelo que entendi... (mais fcil usar o pylab/matplotlib)

```python
#da forma como o curso Let's Code ensina

def get_datasets(y, labels):
  if type(y[0]) == list:
    datasets =[]
    for i in range(len(y)):
      datasets.append({
        'label' : labels[i] ,
        'data' : y[i]
      })
    return datasets
  else:
    return [
            {
                'label' : labels[0],
                'data' : y
            }
    ]
```

```python
#da forma como o curso Let's Code ensina
def set_title(title= ''):
  if title != '':
    display = 'true'
  else:
    display = 'false'
  return {
      'title' : title ,
      'display' : display

  }
```

```python
#da forma como o curso Let's Code ensina
def create_chart(x, y, labels, kind='bar', title=''):

  datasets = get_datasets(y, labels)
  options = set_title(title)

  chart={
      'type' : kind,
      'data' : {
          'labels' : x,
          'datasts' : datasets
      },
      'options' : options
  }

  return chart
```

```python
#da forma como o curso Let's Code ensina
def get_api_chart(chart):
  url_base = 'https://quickchart.io/chart'
  resp = r.get(f'{url_base}?c={str(chart)}')
  return resp.content
```

```python
#da forma como o curso Let's Code ensina
def save_image(path, content):
  with open(path, 'wb') as image:
    image.write(content)
    
```

```python
#da forma como o curso Let's Code ensina
from PIL import Image
from IPython.display import display
```

```python
#da forma como o curso Let's Code ensina
def display_image(path):
  img_pill = Image.open(path)
  display(img_pill)
```

```python
#da forma como o curso Let's Code ensina
y_data_1 = []
for obs in final_data[1::1]:
  y_data_1.append(obs[confirmados])

y_data_2 = []
for obs in final_data[1::1]:
  y_data_2.append(obs[confirmados])

labels =['confirmados' , 'recuperados']

x = []
for obs in final_data[1::10]:
  x.append(obs[data].strftime('%d/%m/%y'))

chart = create_chart(x, [y_data_1 , y_data_2], labels, kind='bar', title='Grfico confirmados X recuperados')
chart_content = get_api_chart(chart)
save_image('grafico.png' , chart_content)
display_image('grafico.png')
```
